```python
import PySimpleGUI as sg

# 设置页面风格和色号
sg.theme('Material2')  # 使用现代的主题风格
sg.theme_background_color('#6BB952')  # 页面底色

# 文本和按钮样式设定
title_font = ('Arial', 22, 'bold')
title_color = '#C4A751'
text_color = '#6BB952'
button_font = ('Arial', 22, 'bold')
button_text_color = '#C4A751'
button_fill_color = '#EC748B'

# 定义 GUI 窗口布局
layout = [
    [sg.Text('Characteristics of biochar', font=title_font, text_color=title_color)],
    [sg.Text('C:', size=(5, 1), text_color=text_color), sg.InputText(key='-C-', size=(10, 1)),
     sg.Text('H/C:', size=(5, 1), text_color=text_color), sg.InputText(key='-HC-', size=(10, 1)),
     sg.Text('(O+N)/C:', size=(9, 1), text_color=text_color), sg.InputText(key='-ONC-', size=(10, 1))],
    [sg.Text('BET:', size=(5, 1), text_color=text_color), sg.InputText(key='-BET-', size=(10, 1)),
     sg.Text('D:', size=(3, 1), text_color=text_color), sg.InputText(key='-D-', size=(10, 1)),
     sg.Text('pH_pzc:', size=(7, 1), text_color=text_color), sg.InputText(key='-pH_pzc-', size=(10, 1))],
    [sg.Text('Adsorption conditions', font=title_font, text_color=title_color)],
    [sg.Text('pH_sol:', size=(7, 1), text_color=text_color), sg.InputText(key='-pH_sol-', size=(10, 1)),
     sg.Text('T:', size=(3, 1), text_color=text_color), sg.InputText(key='-T-', size=(10, 1)),
     sg.Text('C0:', size=(5, 1), text_color=text_color), sg.InputText(key='-C0-', size=(10, 1))],
    [sg.Text('Parameters of dye', font=title_font, text_color=title_color)],
    [sg.Text('A:', size=(3, 1), text_color=text_color), sg.InputText(key='-A-', size=(10, 1)),
     sg.Text('B:', size=(3, 1), text_color=text_color), sg.InputText(key='-B-', size=(10, 1)),
     sg.Text('S:', size=(3, 1), text_color=text_color), sg.InputText(key='-S-', size=(10, 1)),
     sg.Text('V:', size=(3, 1), text_color=text_color), sg.InputText(key='-V-', size=(10, 1))],
    [sg.Button('Output', button_color=(button_text_color, button_fill_color), font=button_font)]
]

# 创建窗口
window = sg.Window('Simple GUI Example', layout, finalize=True)

# 设置整体界面大小以实现长方形页面效果
window.set_min_size((600, 400))

# 事件循环
while True:
    event, values = window.read()
    if event == sg.WINDOW_CLOSED:
        break
    elif event == 'Output':
        # 获取用户输入的值并可以在这里处理
        pass

# 关闭窗口
window.close()
```
